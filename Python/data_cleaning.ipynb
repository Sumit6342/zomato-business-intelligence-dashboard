```python
df.rename(columns={
         'approx_cost(for two people)' : 'approx_cost',
         'listed_in(type)' : 'listed_type',
         'listed_in(city)' : 'listed_city'
} , inplace=True)
```

```python
df.columns
```

```python
df['rate'].unique()
```

```python
df['rate'] = df['rate'].replace('NEW', np.nan)
df['rate'] = df['rate'].replace('-', np.nan)
```

```python
df['rate'] = df['rate'].str.replace('/5','',regex=False)
```

```python
df['rate'] = df['rate'].astype(float)
```

```python
df['rate'].unique()
```

```python
df['approx_cost'].head()
```

```python
df['approx_cost'] = df['approx_cost'].str.replace(',','',regex=False)
```

```python
df['approx_cost'] = df['approx_cost'].astype(float)
```

```python
df['approx_cost'].head()
```

```python
df["rate"] = df["rate"].fillna(df["rate"].median())
df["rest_type"] = df["rest_type"].fillna("Unknown")
df["cuisines"] = df["cuisines"].fillna("Unknown")
df["approx_cost"] = df["approx_cost"].fillna(
    df["approx_cost"].median()
)
df["location"] = df["location"].fillna("Unknown")
df["phone"] = df["phone"].fillna("Not Available")
df["menu_item"] = df["menu_item"].fillna("Not Available")
df["dish_liked"] = df["dish_liked"].fillna("Not Available")
```

```python
df.info()
```

```python
df.to_csv("../Dataset/zomato_cleaned.csv", index=False, encoding="utf-8")
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:@127.0.0.1:3306/zomato_data_analysis"
)
```

```python
from sqlalchemy.dialects.mysql import LONGTEXT, TEXT, VARCHAR

df.to_sql(
    "zomato",
    con=engine,
    if_exists="replace",
    index=False,
    dtype={
        "reviews_list": LONGTEXT(),
        "menu_item": LONGTEXT(),
        "dish_liked": TEXT(),
        "cuisines": VARCHAR(255)
    },
    chunksize=1000,
    method="multi"
)

print("✅ Data Imported Successfully!")

print("✅ Data Imported Successfully!")
```
